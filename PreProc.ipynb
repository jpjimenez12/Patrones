```python
mushroom["Classes"] = mushroom["Classes"].replace({"e" : "Edible", "p" : "Poisonous"})
mushroom["Cap Shape"] = mushroom["Cap Shape"].replace({"b": 0, "c" : 1, "x" : 2, "f" : 3, "k" : 4, "s" : 5})
mushroom["Cap Surface"] = mushroom["Cap Surface"].replace({"f" : 0, "g" : 1, "y" : 2, "s" : 3})
mushroom["Cap Color"] = mushroom["Cap Color"].replace({"n" : 0, "b" : 1, "c" : 2, "g" : 3, "r" : 4, "p" : 5, "u" : 6, "e" : 7, "w" : 8, "y" : 9})
mushroom["Bruises?"] = mushroom["Bruises?"].replace({"f" : 0, "t" : 1})
mushroom["Odor"] = mushroom["Odor"].replace({"n" : 0, "a" : 1, "l" : 2, "c" : 3, "y" : 4, "f" : 5, "m" : 6, "p" : 7, "s" : 8})
mushroom["Gill Attachment"] = mushroom["Gill Attachment"].replace({"n" : 0, "a" : 1, "d" : 2, "f": 3})
mushroom["Gill Spacing"] = mushroom["Gill Spacing"].replace({"c" : 0, "w" : 1, "d" : 2})
mushroom["Gill Size"] = mushroom["Gill Size"].replace({"n" : 0, "b" : 1})
mushroom["Gill Color"] = mushroom["Gill Color"].replace({"k" : 0, "n" : 1, "b" : 2, "h" : 3, "g" : 4, "r" : 5, "o" : 6, "p" : 7, "u" : 8, "e" : 9, "w" : 10, "y" : 11})
mushroom["Stalk Shape"] = mushroom["Stalk Shape"].replace({"e" : 0, "t" : 1})
mushroom["Stalk Root"] = mushroom["Stalk Root"].replace({"b" : 0, "c" : 1, "u" : 2, "e" : 3, "z" : 4, "r" : 5, "?" : np.nan})
mushroom["Stalk Surface Above Ring"] = mushroom["Stalk Surface Above Ring"].replace({"f" : 0, "y" : 1, "k" : 2, "s" : 3})
mushroom["Stalk Surface Below Ring"] = mushroom["Stalk Surface Below Ring"].replace({"f" : 0, "y" : 1, "k" : 2, "s" : 3})
mushroom["Stalk Color Above Ring"] = mushroom["Stalk Color Above Ring"].replace({"n" : 0, "b" : 1, "c" : 2, "g" : 3, "o" : 4, "p" : 5, "e" : 6, "w" : 7, "y" : 8})
mushroom["Stalk Color Below Ring"] = mushroom["Stalk Color Below Ring"].replace({"n" : 0, "b" : 1, "c" : 2, "g" : 3, "o" : 4, "p" : 5, "e" : 6, "w" : 7, "y" : 8})
mushroom["Veil Type"] = mushroom["Veil Type"].replace({"p" : 0, "u" : 1})
mushroom["Veil Color"] = mushroom["Veil Color"].replace({"n" : 0, "o" : 1, "w" : 2, "y" : 3})
mushroom["Ring Number"] = mushroom["Ring Number"].replace({"n" : 0, "o" : 1, "t" : 2})
mushroom["Ring Type"] = mushroom["Ring Type"].replace({"n" : 0, "c" : 1, "e" : 2, "f" : 3, "l" : 4, "p" : 5, "s" : 6, "z" : 7})
mushroom["Spore Print Color"] = mushroom["Spore Print Color"].replace({"k" : 0, "n" : 1, "b" : 2, "h" : 3, "r" : 4, "o" : 5, "u" : 6, "w" : 7, "y" : 8})
mushroom["Population"] = mushroom["Population"].replace({"a" : 0, "c" : 1, "n" : 2, "s" : 3, "v" : 4, "y" : 5})
mushroom["Habitat"] = mushroom["Habitat"].replace({"g" : 0, "l" : 1, "m" : 2, "p" : 3, "u" : 4, "w" : 5, "d" : 6})
```

```python
mushroom.head()
```

Ahora se cambiara el valor NaN que se muestran en la columna "Stalk Root" por la media de los valores de dicha columna.

```python
mushroom["Stalk Root"].fillna(mushroom["Stalk Root"].mean(), inplace = True)
```

```python
mushroom.head()
```

```python
mushroom["Stalk Root"] = mushroom["Stalk Root"].round()
```
